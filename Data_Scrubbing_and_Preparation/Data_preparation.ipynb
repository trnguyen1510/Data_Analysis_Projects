***Explore how the attributes vary or relate to each other. Calculate and visualize correlations using correlation matrix***

```python
#There are some categorical data such as Marital Status, Insurance Type, Injure Type and Overnight hospital stays. We have to convert these Categorical Data to Numerical Data to run the correlation matrix.
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
```

```python
df1['Marital Status'].value_counts()
```

```python
le = LabelEncoder()
```

```python
#Tranforming columns with obj values to numerical values and add those in 
df1['Marital_le'] = le.fit_transform(df1["Marital Status"])
df1['Injury Type Cat'] = le.fit_transform(df1["Injury Type"])
df1['Overnight Stay Cat'] = le.fit_transform(df1["Overnight Hospital Stay"])
df1
```

```python
# Correlations matrix for numerical variables
corr=df1.corr()
corr
```

```python
# Correlation matrix heatmap for numerical variables

ax = sns.heatmap(corr, 
                 vmin=-1, 
                 vmax=1, 
                 center=0,
                 cmap=sns.diverging_palette(20, 220, n=200), 
                 square=True)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

Red means negative and Blue means positive. The stronger the color, the larger the correlation magnitude.
- ID, Num Claimants, Overnight Hospital Stay and Injury Type do not have any relationship with other attributes
- Income of Policy Holder has negative relationship with Marital Status (-0.377974)
- Claim Amount has positive relationships with Claim Amount Received (0.797659) and Revenue (0.797658)
- Total Claimed has positive relationships with Num Claim (0.946770), Num Soft Tissue(0.424877)
- Num Claims has positive relationships with Total Claim(0.946770), Num Soft Tissue (0.491157)
- Num Soft Tissue has positive relationships with Total Claim (0.424877), Num Claims (0.491157), %Soft Tissue(0.816956)
- % Soft Tissue has positive relationships with Num Soft Tissue (0.816956)
- Claim Amount Recieved has positive relationship with Claim Amount (0.797659) and Revenue (1.0), weak negative relationship with Fraud Flag(-0.304245)
- Frau Flag has negative relationship with Claim Amount Received (-0.304245) and Revenue (-0.304252)
- Revenue has postive relationship with Claim Amount (0.797658), Claim Amount Recieved (1.0), and negative relationship with Fraud Flag (-0.304252)

***Eliminating some attributes based on correlation matrix***
